```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('C:\\Users\\Yash\\Documents\\component_image.jpg')
plt.imshow(image)
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.show()
```

```python
threshold_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)[1]
plt.imshow(threshold_image)
plt.show()
```

```python
otsu_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
plt.imshow(otsu_image)
plt.show()
```

```python
adaptive_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 45, 2)
plt.imshow(adaptive_image)
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
closed = cv2.morphologyEx(gray_image, cv2.MORPH_CLOSE, kernel)
opened = cv2.morphologyEx(closed, cv2.MORPH_OPEN, kernel)
closed_image = cv2.adaptiveThreshold(closed, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 45, 2)
opened_image = cv2.adaptiveThreshold(opened, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 45, 2)
print("Opened Image")
plt.imshow(opened_image)
plt.show()
print("Closed Image")
plt.imshow(closed_image)
plt.show()
```
